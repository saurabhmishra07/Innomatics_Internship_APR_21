```python
def mutate_string(string, position, character):
    l = list(string)
    l[position] = character
    string = ''.join(l)
    return string


if __name__ == '__main__':
    s = input("Enter the characters")
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)       
            
        





```

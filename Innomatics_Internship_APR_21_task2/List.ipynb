```python
if __name__ == '__main__':
    N = int(input("Enter the number"))
    L=[]
    for i in range(0,N):
        a=input("Enter the number").split()
        if a[0] == "insert":
            L.insert(int(a[1]),int(a[2]))
        elif a[0] == "append":
            L.append(int(a[1]))
        elif a[0] == "pop":
            L.pop();
        elif a[0] == "print":
            print(L)
        elif a[0] == "remove":
            L.remove(int(a[1]))
        elif a[0] == "sort":
            L.sort();
        else:
            L.reverse()

        
    
```

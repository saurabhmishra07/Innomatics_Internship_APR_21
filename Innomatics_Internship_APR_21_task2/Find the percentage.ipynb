```python
if __name__ == '__main__':
    n = int(input("enter the number"))
    student_marks = {}
    for i in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    scores2 = student_marks[query_name]
    print("{0:.2f}".format(sum(scores2)/(len(scores2))))


        
    
```
